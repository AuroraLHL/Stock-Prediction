```python
import yfinance as yf  #需要挂梯子
yf.pdr_override()
import os
from datetime import datetime
```

```python
# get stock data
def get_stock_data(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    return stock_data

tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN','600967.SS'] 
start_date = datetime(2018,12,1)
end_date = datetime(2023,12,15)
directory = 'StockData'

if not os.path.exists(directory): # create directory if it doesn't exist
    os.makedirs(directory)

for ticker in tickers:
    data = get_stock_data(ticker, start_date, end_date)
    data.to_csv(f'{directory}/{ticker}.csv') # save data to csv file
    
```
