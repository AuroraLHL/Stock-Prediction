```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
from datetime import datetime
```

# 1 爬取多只股票，并观察走势

```python
company = ['AAPL','DELL','JPM','C','BAC']
startdate = datetime(2022,1,1)
enddate = datetime(2023,2,1)
top_tech_df = pdr.get_data_yahoo(company, start=startdate, end=enddate)['Close']
top_tech_df.head(3)
```

```python
plt.rcParams["figure.dpi"]=150
top_tech_df.plot(figsize=(15,8))
plt.title("多只股票走势图")
plt.show()
```

# 2相关关系图

```python
corr= top_tech_df.corr(method = 'pearson', min_periods = 1)#pearson方法计算相关性
# 设置Axes的标题
f, ax= plt.subplots(dpi=100)
sns.heatmap(corr,cmap='RdBu', linewidths = 0.05,annot=True, ax = ax)
ax.set_title('Correlation between features')
plt.show()
plt.close()
```
